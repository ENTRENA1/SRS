```python
b = 1234513614970965380450347867786657369506079072062126064869919324956143758812637118581694154929099396752251787268346548051895320171079663652680741564200286881487888963198953533111702360348366584491871177238206448551840553059455017102276155580936577819310963989369822041154857860188417712902205755086669022305216052360483623367597150425938247630127368253363295292024736143937779912318142315499711747531882501424082252281646411119545875582301121408132266980986547390256366071064252128124210381555015623700519223183615506726230814115479519473583475357010445966332533796030494190611947618181858300094662765895526963615406
B = pow(g, b, p)
print('B = ' + str(B))
```

`Bob` calcula la clave K:
$$K_{bob} = A^b mod p $$

```python
K_bob = pow(A,b,p)
print('Clave de Bob = ' + str(K_bob))
```

`Alice` recibe B de `Bob` y calcula la clave K:
$$K_{alice} = B^a mod p $$

```python
K_alice = pow(B,a,p)
print('Clave de Alice = ' + str(K_alice))
```

`Alice` y `Bob` tienen la misma clave K:

```python
if K_alice == K_bob:
    print('Alice y Bob tienen la misma clave')
else:
    print('Algo ha ido mal')
```
